```python
import pandas as pd

from src.auth import typesense_client
```

```python
COLLECTION_NAME = "awesome_selfhosted_repos"

dataframe = pd.read_json(
  typesense_client.collections[COLLECTION_NAME].documents.export(),
  lines=True
)

dataframe = dataframe.set_index('id')
```

```python
# Projects Sorted by stars
dataframe.sort_values(by=['stars'], ascending=False).head(20)
```

```python
# see how many project were written with which language
result = dataframe.groupby(['language']).agg({'language': 'count', 'stars': 'sum'})
result = result.rename(columns={'language': 'count'})
result['avg_stars'] = result['stars'] / result['count']

result = result.sort_values(by='avg_stars', ascending=False)

result
```
