```python
import os

def find_files(directory):
    file_list = []
    
    for root, _, files in os.walk(directory):
        for file in files:
            file_list.append(os.path.join(root, file))
    
    return file_list
```

```python
import markdown
import re
import yaml

def extract_frontmatter_from_markdown_file(file_path):
    frontmatter_pattern = re.compile(r'^---\s*\n(.*?)\n---\s*\n', re.MULTILINE | re.DOTALL)

    with open(file_path, 'r', encoding='utf-8') as file:
        md_content = file.read()

        # Extract frontmatter using a regular expression
        frontmatter_match = frontmatter_pattern.match(md_content)

        if frontmatter_match:
            frontmatter_text = frontmatter_match.group(1)
            return yaml.safe_load(frontmatter_text)
            
        else:
            return None

```

```python
# Example usage:
markdown_file = "/Users/u1082926/obsidian/sources/Readwise/The Sixth Extinction.md"

extract_frontmatter_from_markdown_file(markdown_file)
```

```python
def is_markdown_file(file_path):
    # Get the file extension
    file_extension = os.path.splitext(file_path)[1].lower()

    # Check if the file extension is .md or .markdown
    return file_extension in ['.md', '.markdown']
```

```python
import yaml

def is_valid_yaml(yaml_string):
    try:
        yaml.safe_load(yaml_string)
        return True
    except yaml.YAMLError:
        return False

yaml_string = input("What Yaml to add?: ")

if is_valid_yaml(yaml_string):
    print("The string is valid YAML.")
else:
    print("The string is not valid YAML.")
```
